### DATA HANDLING

```python
a = %sql SELECT * FROM conflicts WHERE jour = '2022-8-19';
df = pd.DataFrame(a)
```

```python
x = df['conflit']
y = df['ponderation']
```

### PLOTTING

```python
plt.style.use('classic')

slices = y
labels = x

plt.pie(slices, labels=labels, startangle=90, wedgeprops={'edgecolor': 'black'}, autopct='%1.1f%%')

plt.tight_layout()
plt.title('Problèmes du jour')
plt.show()
```
